```python
def sumArray(arr):
    l = len(arr)
    if l == 0:
        return 0
    remainingArray = arr[1:]
    nextSum = arr[0] + sumArray(remainingArray)
    
    return nextSum
```

```python
arr = [9,8,62,34,56]
print(sumArray(arr))
```

```python
def checkNumber(arr,x):
    l = len(arr)
    if l == 0:
        return False
    if x == arr[0]:
        return True
    remainingArray = arr[1:]
    if x in remainingArray:
        return True
    else:
        return False
    checkNumber(remainingArray, x)
```

```python
arr = [1,2,3,4,5]
print(checkNumber(arr,1))
```

```python
def replaceChar(s,a,b):
    if len(s) == 0:
        return s
    smallerOutput = replaceChar(s[1:] , a, b)
    if s[0] == a:
        return b + smallerOutput
    else:
        return s[0] + smallerOutput
```

```python
def partition(a, si, ei):
    
    pivot = a[si]
    count = 0
    #Find no of elements smaller than pivot
    for i in range(si, ei+1):
        if a[i] < pivot:
            count = count + 1
            
    a[si+count] , a[si] = a[si], a[si+count] #Put the first element in the correct place.
    pivot = si+count
    
    #Putting elements smaller than pivot on the left of pivot and elements which are greater than pivot on the right of pivot
    
    i = si
    j = ei
    
    while i < j: #When i becomes equal to j that means both of them are pointing to the same elemnt that is the pivot
        if a[i] < pivot:
            i = i + 1
        elif a[j] >= pivot:
            j = j - 1
        else:
            a[i] , a[j] = a[j], a[i]
            i = i + 1
            j = j - 1
            
    return pivot
    
```

```python
def staircase(n):
    if n == 0:
        return 1
    x = staircase(n-1)
    y = staircase(n-2)
    z = staircase(n-3)
    
    return x + y + z
    
```

```python
print(staircase(4))
```
